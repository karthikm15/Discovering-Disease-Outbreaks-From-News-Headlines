```python
import pandas as pd
import geonamescache
import re
import unidecode


file = open('headlines.txt', 'r')

arr = []
while (True):
    if (file.readline() == ''):
        break
    else:
        arr.append(file.readline()[:-1])

       
for i in arr:
    re.sub(r'\W+', '', i)
   
arr = [unidecode.unidecode(i) for i in arr]
dictionary = {'headline':[], 'city':[], 'country':[]}


gnc = geonamescache.GeonamesCache()
country_names = [k for k in gnc.get_countries_by_names()]

#print(arr)
for i in arr:
    dictionary['headline'].append(i)
    words = i.split()
    for x in range(len(words)):
        if (not words[x][-1].isalnum()):
            words[x] = words[x][:-1]
    for j in range(len(words)):
        poss_cities = []
        string = words[j]
        if (gnc.get_cities_by_name(words[j]) != []):
            poss_cities.append(string)
       
        for k in range(j+1, len(words)):
            string += " " + words[k]
            if (gnc.get_cities_by_name(string) != []):
                poss_cities.append(string)
       
        if (len(poss_cities) != 0):
            dictionary['city'].append(poss_cities[-1])
            break
    for j in range(len(words)):
        poss_countries = []
        string = words[j]
        if (string in country_names):
            poss_countries.append(string)
       
        for k in range(j+1, len(words)):
            string += " " + words[k]
            if (string in country_names):
                poss_countries.append(string)
       
        if (len(poss_countries) != 0):
            dictionary['country'].append(poss_countries[-1])
            break
    if (len(dictionary['headline']) != len(dictionary['country'])):
        dictionary['country'].append('')
    if (len(dictionary['headline']) != len(dictionary['city'])):
        dictionary['city'].append('')

df = pd.DataFrame(data = dictionary)
df
```
